## Spatial Filtering

```python
import numpy as np
from osgeo import gdal
import matplotlib.pyplot as plt


def load_raster_band (img_file, band = 1):
    """band is one based"""
    dataset = gdal.Open(img_file, gdal.GA_ReadOnly)
    return dataset.GetRasterBand(band).ReadAsArray()

def show_img(img,cmap='gray'):
    fig=plt.figure(figsize=(8, 6))
    plt.imshow(img, cmap=cmap)
    plt.colorbar()
    plt.show()

from skimage import data, filters
from copy import deepcopy
```

## gaussian smoothing and edge detection by subtraction

skimage.filters.gaussian(image, sigma=1, ...


```python
print('original')
img = data.camera()
show_img(img)

print('some smothing')
gaus_img = filters.gaussian(img,sigma=1)
show_img(gaus_img)

print('more smothing')
gaus_img = filters.gaussian(img,sigma=2)
show_img(gaus_img)

print('edges')
edges = gaus_img - filters.gaussian(img,sigma=1)
show_img(edges)
```

## Rank Filters

```python
from skimage import morphology
print('original x2 to make some noise')
img = data.moon() *2
show_img(img)

print('median')
filter_img = filters.rank.median(img, selem=morphology.disk(3))
show_img(filter_img)

print('minimum')
filter_img = filters.rank.minimum(img, selem=morphology.disk(3))
show_img(filter_img)

print('maximum')
filter_img = filters.rank.maximum(img, selem=morphology.disk(3))
show_img(filter_img)
```

## Laplacian

skimage.filters.laplace(image, ksize=3, mask=None)

```python

print('original blured')
img = filters.gaussian(data.text(),3)
show_img(img)

print('Laplacian')
lp = filters.laplace(img, 5)
show_img(lp)

print('Sharpened')

show_img(lp+img)
```

## Edge Detection


```python
print('original blured')
img = filters.gaussian(data.camera(),3)
show_img(img)

print('Laplacian')
lp = filters.laplace(img, 5)
show_img(lp)

print('Roberts')
lp = filters.roberts(img)
show_img(lp)

print('Sobel')
lp = filters.sobel(img)
show_img(lp)

print('Sobel for vertical edges')
lp = filters.sobel_v(img)
show_img(lp)
```
